```
import torch
from transformers import pipeline,AutoTokenizer,AutoModelForQuestionAnswering
```

```
model_name = "distilbert-base-uncased-distilled-squad"  # Model name for question answering
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
nlp = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```
context = "The discovery of electricity revolutionized the world. It led to the development of various inventions and technologies that transformed the way we live. One of the key figures in the field of electricity is Thomas Edison, who is credited with inventing the practical electric light bulb."
```

```
question = "Who is credited with inventing the electric light bulb?"
```

```
result = nlp(question=question, context=context)
```

```
answer = result["answer"]
start_pos = result["start"]
end_pos = result["end"]
```
